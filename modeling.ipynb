```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import os
```

```python
df = pd.read_csv(os.getcwd() + '\\data\\kaggle_survey_2020_responses.csv').iloc[: , 1:]
np.random.seed(0)

df = df[~df.Q5.isin(['Student', 'Currently not employed'])]
df = df[~df.Q2.isin(['Nonbinary', 'Prefer not to say', 'Prefer to self-describe'])]
df['Q3'] = df['Q3'].str.replace('United States of America','USA')
df['Q3'] = df['Q3'].str.replace('United Kingdom of Great Britain and Northern Ireland','UK')
df = df.dropna(subset=['Q24'])

def transform_salary_range (value):
    a = ''
    for e in value:
        if e == '0' or e == '1' or e == '2' or e == '3' or e == '4' or e == '5' or e == '6' or e == '7' or e == '8' or e == '9' or e == '-':
            a = a + e
    if '-' in a:
        a = a.split('-')
        return float((int(a[1])+int(a[0]))/2)
    if a == '':
        return value
    else: 
        return float(a)

df.Q24 = df.Q24.apply(transform_salary_range)

df.head()
```
